```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
# Coordinates for launch sites
launch_sites = {
    "CCAFS LC-40": (28.562302, -80.577356),
    "CCAFS SLC-40": (28.563197, -80.576820),
    "KSC LC-39A": (28.573255, -80.646895),
    "VAFB SLC-4E": (34.632834, -120.610745)
}

# Corresponding nearest coastline points (you collected these with MousePosition)
coastlines = {
    "CCAFS LC-40": (28.56146, -80.56824),
    "CCAFS SLC-40": (28.56146, -80.56824),
    "KSC LC-39A": (28.56146, -80.56824),
    "VAFB SLC-4E": (34.63434, -120.62611)
}

# Compute and print the distances
for site in launch_sites:
    lat1, lon1 = launch_sites[site]
    lat2, lon2 = coastlines[site]
    distance = calculate_distance(lat1, lon1, lat2, lon2)
    print(f"Distance from {site} to coastline: {distance:.2f} km")
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
# Define proximity coordinates for each launch site
# Format: site_name: (city_coords, highway_coords, railroad_coords)

proximities = {
    "CCAFS LC-40": {
        "city": (28.61105, -80.80056),
        "highway": (28.56183, -80.57053),
        "railroad": (28.57394, -80.65421)
    },
    "KSC LC-39A": {
        "city": (28.61105, -80.80056),
        "highway": (28.57834, -80.65557),
        "railroad": (28.57394, -80.65421)
    },
    "VAFB SLC-4E": {
        "city": (34.63632, -120.45513),
        "highway": (34.65877, -120.5437),
        "railroad": (34.63425, -120.62463)
    }
}

# Loop through each launch site
for site in spacex_df['Launch Site'].unique():
    # Skip if the site isn't in your proximity dictionary
    if site not in proximities:
        continue

    # Get launch site lat/lon from df
    site_row = spacex_df[spacex_df['Launch Site'] == site].iloc[0]
    launch_lat = site_row['Lat']
    launch_lon = site_row['Long']

    # Loop through each proximity type
    for label, (prox_lat, prox_lon) in proximities[site].items():
        # Calculate distance
        distance = calculate_distance(launch_lat, launch_lon, prox_lat, prox_lon)

        # Add a marker for the proximity point
        folium.Marker(
            location=[prox_lat, prox_lon],
            icon=folium.DivIcon(
                icon_size=(150, 36),
                icon_anchor=(0, 0),
                html=f'<div style="font-size: 11px; color:#4a148c;"><b>{label.title()}: {distance:.2f} km</b></div>'
            )
        ).add_to(site_map)

        # Draw a line to the launch site
        folium.PolyLine(
            locations=[[launch_lat, launch_lon], [prox_lat, prox_lon]],
            weight=2,
            color='purple',
            tooltip=f"{site} to {label.title()}"
        ).add_to(site_map)

site_map
```
