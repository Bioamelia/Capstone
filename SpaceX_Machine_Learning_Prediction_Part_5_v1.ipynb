This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix');
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed'])
    plt.show()
```

```python
from sklearn.tree import _tree

def extract_success_paths(tree, feature_names, class_index=1):
    """
    Extracts decision rules from a fitted DecisionTreeClassifier that lead to pure success.

    Parameters:
        tree: trained DecisionTreeClassifier
        feature_names: list of column names
        class_index: index of 'Success' class in value array (default 1 if classes = ['Failure', 'Success'])

    Returns:
        List of rule strings that lead to 100% success
    """
    tree_ = tree.tree_
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]

    success_paths = []

    def recurse(node, path):
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]

            # LEFT branch: feature <= threshold
            recurse(tree_.children_left[node], path + [f"{name} <= {threshold:.3f}"])

            # RIGHT branch: feature > threshold
            recurse(tree_.children_right[node], path + [f"{name} > {threshold:.3f}"])
        else:
            values = tree_.value[node][0]
            total = sum(values)
            predicted_class = values.argmax()
            # Check for pure success
            if predicted_class == class_index and values[class_index] == total:
                success_paths.append(path)

    recurse(0, [])
    return success_paths
```

```python
# First, ensure you have feature names and the best estimator:
feature_names = X_df.columns.tolist()  # before scaling
best_tree = tree_cv.best_estimator_

# Now extract success paths
success_rules = extract_success_paths(best_tree, feature_names)

# Print them
for i, rule in enumerate(success_rules, 1):
    print(f"Success Rule {i}:")
    for step in rule:
        print("   ", step)
    print()
```

```python
def interpret_rule_step(step):
    """
    Convert threshold conditions on binary one-hot encoded features into readable logic.
    """
    import re
    match = re.match(r"(.*) (<=|>) ([-+]?[0-9]*\.?[0-9]+)", step)
    if not match:
        return step  # fallback

    feature, operator, threshold = match.groups()
    threshold = float(threshold)

    # If it's a one-hot feature, threshold ~0.5 indicates True/False logic
    if '_True' in feature or '_False' in feature:
        base = feature.replace('_True', '').replace('_False', '')
        if '_True' in feature:
            return f"{base} is {operator} True" if threshold < 0.5 else f"{base} is {operator} False"
        else:
            return f"{base} is {operator} False" if threshold < 0.5 else f"{base} is {operator} True"

    return step  # for numerical features

for i, rule in enumerate(success_rules, 1):
    print(f"Success Rule {i}:")
    for step in rule:
        print("   ", interpret_rule_step(step))
    print()
```
